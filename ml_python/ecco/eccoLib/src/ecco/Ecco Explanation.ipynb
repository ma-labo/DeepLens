```python
import ecco
```

```python
text = '''

Bicycle Technologies International Ltd. (BTI) is a global distributor of bicycle components and accessories based in Santa Fe NM USA. BTI also participates in the community through cycling advocacy on local and national levels. The company is privately held and does not disclose its sales figures.
'''
lm = ecco.from_pretrained('distilbert-base-uncased', activations=True)
inputs = lm.tokenizer([text], return_tensors="pt")
output = lm(inputs)
nmf_1 = output.run_nmf(n_components=8) 
nmf_1.explore()
```

```python
nmf_1.explore(filter_token=True)
```

```python
nmf_1.explore(filter_token=True, top_k=10)
```

```python
def ecco_saliency(text, language_model='distilbert-base-uncased', n_components=8, filter_token=False, top_k=-1):
    lm = ecco.from_pretrained(language_model, activations=True)
    if 'bert' in language_model:
        inputs = lm.tokenizer([text], return_tensors="pt")
        output = lm(inputs)
    else:
        output = lm.generate(text, generate=1, do_sample=True)
    nmf_1 = output.run_nmf(n_components=n_components) 
    nmf_1.explore(filter_token=filter_token, top_k=top_k)
```

```python
text = '''
Tsuga mertensiana known as mountain hemlock is a species of hemlock native to the west coast of North America with its northwestern limit on the Kenai Peninsula Alaska and its southeastern limit in northern Tulare County California.
'''
ecco_saliency(text)
```

```python
ecco_saliency(text, filter_token=True)
```

```python
ecco_saliency(text, filter_token=True, top_k=5)
```
